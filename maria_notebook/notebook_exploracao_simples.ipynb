# Exploração de Dados - Transações

Análise inicial do dataset de transações

```python
import pandas as pd
import numpy as np
```

```python
# Carregar dados
df = pd.read_csv('data/transacoes.csv')
print(f"Dataset carregado: {len(df)} linhas")
```

```python
# Ver primeiras linhas
df.head()
```

```python
# Informações do dataset
df.info()
```

```python
# Estatísticas descritivas
df.describe()
```

```python
# Verificar valores nulos
df.isnull().sum()
```

```python
# Distribuição de categorias
df['categoria'].value_counts()
```

```python
# Limpar dados
df_limpo = df.dropna()
df_limpo = df_limpo.drop_duplicates()
print(f"Após limpeza: {len(df_limpo)} linhas")
```

```python
# Criar feature derivada
df_limpo['valor_log'] = np.log1p(df_limpo['valor'])
df_limpo.head()
```

```python
# Salvar dados processados
df_limpo.to_csv('data/transacoes_processado.csv', index=False)
print("Dados processados salvos!")
```
