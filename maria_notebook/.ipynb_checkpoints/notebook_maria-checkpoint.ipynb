```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, classification_report
import numpy as np
```

```python
# Ler dados
df = pd.read_csv('/Users/maria/Desktop/dados_treino_FINAL_v3.csv')
```

```python
# Preparar features
X = df.drop('is_fraud', axis=1)
y = df['is_fraud']
```

```python
# Limpeza (adicionei depois)
df = df.dropna()
df = df.drop_duplicates()
```

```python
# Converter categoria para número
categoria_map = {'A': 0, 'B': 1, 'C': 2, 'D': 3}
X['categoria'] = X['categoria'].map(categoria_map)
```

```python
# Dividir treino/teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Treinar modelo
print("Treinando modelo...")
model = RandomForestClassifier(n_estimators=20, max_depth=8, random_state=42)
model.fit(X_train, y_train)
```

```python
# Avaliar
y_pred = model.predict(X_test)
f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1:.4f}")
```

```python
print(classification_report(y_test, y_pred))
```

```python
# Salvar modelo (ou não? esqueci onde...)
import pickle
with open('modelo_fraude.pkl', 'wb') as f:
    pickle.dump(model, f)
```

```python
# Teste rápido com dados novos
exemplo = pd.DataFrame({
    'id': [99999],
    'valor': [15000.00],
    'hora': [23],
    'categoria': [2]
})
```

```python
pred = model.predict(exemplo)
print(f"Predição: {pred}")
if pred[0] == 1:
    print("FRAUDE DETECTADA!")
else:
    print("Transação legítima")
```
